```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

# Data Loading

```python
train = pd.read_csv('train.csv', index_col='id')
test = pd.read_csv('test.csv', index_col='id')
ss = pd.read_csv('sample_submission.csv')
```

# Combine Train and Test Data

```python
train1 = train.drop('yield', axis=1).copy()
```

```python
train1 = pd.concat([train1, test])
```

# Feature Engineering

```python
train2 = train1.copy()
```

```python
# Based on the correlation matrix, these parameters are highly correlated and have the same correlation with the target so it is safe to remove them
# train2 = train2.drop(['MinOfUpperTRange', 'AverageOfUpperTRange', 'MaxOfLowerTRange', 'MinOfLowerTRange', 'AverageOfLowerTRange', 'RainingDays'], axis=1)
```

# Feature Scaling

```python
train3 = train2.copy()
```

```python
sc = StandardScaler()
train3 = pd.DataFrame(sc.fit_transform(train3), columns=train3.columns)
```

# Target Transformation

```python
y = train['yield']
# sns.histplot(y)
# plt.show()
```

# Split Train and Test data, transform

```python
X = train3.loc[:train.index.max(), :].copy()
test_transformed = train3.iloc[train.index.max()+1:, :]
```

# Base Model

```python
# Define the linear regression model
model = RandomForestRegressor()

# Define the initial set of features to consider
features = X.columns.to_list()

# Define the minimum MAE threshold
min_mae = np.inf

# Iterate over the features and remove the one that results in the lowest MAE
while len(features) > 1:
    mae_values = []
    for feature in features:
        temp_features = features.copy()
        temp_features.remove(feature)
        model.fit(X[temp_features], y)
        y_pred = model.predict(X[temp_features])
        mae = mean_absolute_error(y, y_pred)
        mae_values.append(mae)
    best_feature_index = np.argmin(mae_values)
    best_feature_mae = mae_values[best_feature_index]
    if best_feature_mae < min_mae:
        min_mae = best_feature_mae
        features.pop(best_feature_index)
    else:
        break

# Print the best set of features
print(features)
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=0)

lr = LinearRegression()

results = []

results.append(-cross_val_score(lr, X[features], y, cv=cv, scoring='neg_mean_absolute_error'))

print(f'MAE: {np.mean(results).round(2)}')
```

```python
cv = KFold(n_splits=10, shuffle=True, random_state=0)

rf = RandomForestRegressor()

results = []

results.append(-cross_val_score(rf, X[features], y, cv=cv, scoring='neg_mean_absolute_error'))

print(f'MAE: {np.mean(results).round(2)}')
```

```python
lr.fit(X, y)
```

```python
rf.fit(X, y)
```

```python
importances = lr.coef_
indices = np.argsort(importances)
plt.figure(figsize = (8, 4))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color = 'violet', align = 'center')
plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)
plt.figure(figsize = (8, 4))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color = 'violet', align = 'center')
plt.yticks(range(len(indices)), [X.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

# Training

```python
rf.fit(X, y)
```

# Prediction

```python
y_pred = rf.predict(test_transformed)
```

# Submission

```python
submission = pd.DataFrame(pd.concat([pd.Series(test_transformed.index, name='id'), pd.Series(y_pred, name='yield')], axis=1))
```

```python
submission.to_csv('base_model_rf.csv', index=False)
```
